```python
import os
from os import listdir
from os.path import isfile, join
from os import walk

def upload_images(path):
    for i in listdir(path):
        append1 = path + '/' + i + '/'
        for j in listdir(append1):
            append2 = path + '/' + i + '/' + j + '/'
            for k in listdir(append2):
                append3 = path + '/' + i + '/' + j + '/' + k
                for l in listdir(append3):
                    append4 = path + '/' + i + '/' + j + '/' + k + '/' + l
                    for m in listdir(append4):
                        append5 = path + '/' + i + '/' + j + '/' + k + '/' + l + '/' + m
                        for n in listdir(append5):
                            append6 = path + '/' + i + '/' + j + '/' + k + '/' + l + '/' + m + '/' + n
                            locatn = '/'+ i + '/' + j + '/' + k + '/' + l + '/' + m + '/' + n
                            boto3.Session().resource('s3').Bucket(bucket).Object(prefix+locatn).upload_file(append6)
```

```python
from shutil import copyfile
import os
from os import listdir
from os.path import isfile, join
from os import walk
```

```python
def move_images():
    count = 0
    for row in range(0,402760):
        partition = df['partition'][row]
        id1 = df['id'][row]
        id2 = df['image_id'][row]
        
        src = 'path_to_dataset/images/' + partition + '/'  +id2[0] + '/' + id2[1] + '/'  + id2[2] + '/'  + id2[3] + '/'  + id2
        dst = partition + '/' + id1 + "_" +id2
        
        copyfile(src, dst)
        count = count + 1
        if(count % 1000 == 0):
            print('moved ----'+ str(count))                        
```

```python
move_images()
```

```python
from PIL import Image 
import os, sys
```

```python
def resize(path):
    count = 0
    dirs = os.listdir( path )
    for item in dirs:
        if os.path.isfile(path + '/' + item):
            im = Image.open(path + '/' + item)
            imResize = im.resize((256,256), Image.ANTIALIAS)
            imResize.save(path + '1/' + item, 'JPEG', quality=90)
            count = count+1
            if(count%1000 == 0):
                print('resized --- ' + str(count))
    print('total -- ' + str(count) + 'completed!')
```

```python
resize('train')
```

```python
resize('val')
```

```python
resize('test')
```
