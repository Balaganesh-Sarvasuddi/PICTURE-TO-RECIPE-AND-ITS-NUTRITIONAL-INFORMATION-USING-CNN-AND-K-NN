### 1.1 Collecting Common Data from layer1 and layer2

```python
import pandas as pd
import numpy as np
import os
```

```python
layer1 = pd.read_json("data/layer1.json")
layer2 = pd.read_json("data/layer2.json")

#Shows the breakdown of recipes in layer1
layer1.groupby(['partition']).count()
```

```python
#Filter layer 1 for only the recipes in Recipe 1M (Currently showing all the recipes in Recipe 1M+)
layer2ids = layer2['id']
common_data = layer1.loc[layer1['id'].isin(layer2ids)]

common_data.to_csv('data/common_data.csv',index = False)
common_data.head(10)
```

### 1.2 merged the requried data into reqired format

```python
required_data=[]
count = 0
for row in range(0,402760):
    for item in layer2['images'][row]:
        required_data.append([layer2['id'][row], item['id']])
    count = count+1
    if(count%1000 == 0):
        print(str(count)+ " -- done!")
print('total '+str(count)+' completed!')
```

```python
df = pd.DataFrame(required_data,columns=
                 ['id', 'image_id'])
df.to_csv('data/required_data.csv',index = False)
df
```

```python
del(layer1)
del(layer2)
```

## 1.3 collecting unque_data

    -since many images have same ingredient informatio

```python
layer2 = pd.read_json("data/layer2.json")
required_data2=[]
count = 0
for row in range(0,402760):
    required_data2.append([layer2['id'][row], layer2['images'][row][0]['id'], layer2['images'][row][0]['url']])
    count = count+1
    if(count%1000 == 0):
        print(str(count)+ " -- done!")
print('total '+str(count)+' completed!')
```

```python
df = pd.DataFrame(required_data2,columns = ['id', 'image_id', 'url'])
df.to_csv('data/required_unique_data.csv',index = False)
df
```
